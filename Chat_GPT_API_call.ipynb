### **Text generation:**

Method 1: using OpenAI module

```python
import openai
import pandas as pd
# Define OpenAI API key 
openai.api_key = "YOUR_API_KEY"
# Set up the model and prompt
model_engine = "text-davinci-003"
prompt = "Once upon a time, in a land far, far away, there was a princess who..."
# Generate a response
completion = openai.Completion.create(
    engine=model_engine,
    prompt=prompt,
    max_tokens=1024,
    n=3,
    stop=None,
    temperature=0.5,
)
response = completion.choices
pd.DataFrame(response,columns=["text"])
```

Method 2: using requests module

```python
import requests

prompt = "Once upon a time, in a land far, far away, there was a princess who..."

parameters = {
            "model": "text-davinci-003",
            "prompt":prompt,
            "max_tokens":1024,
            "n":3,
            "stop":"None",
            "temperature":0.5,
        }

response = requests.post("https://api.openai.com/v1/completions",
           headers = {
              "Content-Type": "application/json",
              "Authorization": "Bearer YOUR_API_KEY"
          },
          json=parameters).json()

pd.DataFrame(response['choices'],columns=["text"])
```

## **Chat Bot:**

```python
import requests

usermessage = "query to select first 10 records for different sql db's !"

parameters = {
  "model": "gpt-3.5-turbo",
  "messages": [{"role": "user", "content": usermessage}],
  "max_tokens":1024,
  "n":3,
  "stop":"None",
  "temperature":0.5,
}

response = requests.post("https://api.openai.com/v1/chat/completions",
           headers = {
              "Content-Type": "application/json",
              "Authorization": "Bearer YOUR_API_KEY"
          },
          json=parameters).json()


df = pd.DataFrame(response['choices'],columns=["message"])
pd.DataFrame(df.message.apply(pd.Series)['content'])
```
